```python
# as this is classification problem. firstly, let's import logistic regression first
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
```

```python
lr=LogisticRegression()
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
dtc= DecisionTreeClassifier()
rfc=RandomForestClassifier()
svc=SVC()
```

```python
def algo_state (x,y,algo):
    max_acc=0
    max_i=0
    for i in range(100):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = i)
        algo.fit(x_train,y_train)
        y_pred = algo.predict(x_test)
        acc= accuracy_score(y_test,y_pred)
        if acc> max_acc:
            max_acc=acc
            max_i = i
            i+=1
    print(f"for algo {algo}, the maximum accuracy is {max_acc} at random_state {max_i}") 
 
```

```python
algo_state(X,Y,lr)
```

```python
algo_state(X,Y,dtc)
```

```python
algo_state(X,Y,rfc)
```

```python
algo_state(X,Y,svc)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# so, from above, rfc seems to be the best model at random_state 91. let's cross check it by cvs
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state = 91)
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
acc= accuracy_score(y_test,y_pred)
acc
```

```python
(cross_val_score(rfc,X,Y,cv=5)).mean()
```
