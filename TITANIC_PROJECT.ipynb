```python
x=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
from sklearn.metrics import accuracy_score,classification_report,roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
lr=LogisticRegression()
dtc=DecisionTreeClassifier()
rfc= RandomForestClassifier()
abc=AdaBoostClassifier()
from sklearn.model_selection import cross_val_score
```

```python
def algo_check (x,y,algo):
    max_acc=0
    max_i=0
    for i in range(100):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = i)
        algo.fit(x_train,y_train)
        y_pred = algo.predict(x_test)
        acc= accuracy_score(y_test,y_pred)
        if acc> max_acc:
            max_acc=acc
            max_i = i
            i+=1
    print(f"for algo {algo}, the maximum accuracy is {max_acc} at random_state {max_i}") 
    
```

```python
algo_check(x,y,lr)
```

```python
algo_check (x,y,dtc)
```

```python
algo_check (x,y,rfc)
```

```python
algo_check (x,y,abc)
```

```python
# it seems that lr is the best algo at random_state 6 among all the above checked algos. let's verify its overfitting.
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = 6)
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
cross_val_score(lr,x,y,cv=7,scoring='accuracy')
```

```python
# hence, not overfitted.
```

```python
# let's hypertune.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'penalty':['l1', 'l2','elasticnet'],
          'fit_intercept':[True,False],
          'intercept_scaling':[1,2,3],
          'solver':['newton-cg','lbfgs','sag'],
          'max_iter':[100,150,200]}
grid=GridSearchCV(lr,param_grid = params,n_jobs=-1)
grid.fit(x_train,y_train)
print(grid.best_params_)
```

```python
lr1=LogisticRegression(fit_intercept=True,intercept_scaling=1,max_iter=100,penalty='l2',solver='newton-cg')
lr1.fit(x_train,y_train)
y_pred=lr1.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
# not significant improvement. so, can take either lr or lr1 as final algo. let's take lr1 as final algo.
```
