```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
lr=LinearRegression()
dtr=DecisionTreeRegressor()
rfr= RandomForestRegressor()
svr=SVR()
from sklearn.model_selection import cross_val_score
```

```python
df_train1
```

```python
df_train1['Item_Weight']=scaler.fit_transform(df_train1[['Item_Weight']])
df_train1['Item_Visibility']=scaler.fit_transform(df_train1[['Item_Visibility']])
df_train1['Item_MRP']=scaler.fit_transform(df_train1[['Item_MRP']])
df_train1['Outlet_Establishment_Year']=scaler.fit_transform(df_train1[['Outlet_Establishment_Year']])
df_train1  
```

```python
x=df_train1.drop('Item_Outlet_Sales',axis=1)
y=df_train1.Item_Outlet_Sales
```

```python
def algo_check (x,y,algo):
    max_acc=0
    max_i=0
    for i in range(100):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state = i)
        algo.fit(x_train,y_train)
        y_pred = algo.predict(x_test)
        acc= r2_score(y_test,y_pred)
        if acc> max_acc:
            max_acc=acc
            max_i = i
            i+=1
    print(f"for algo {algo}, the maximum accuracy is {max_acc} at random_state {max_i}") 
```

```python
algo_check (x,y,lr)
```

```python
algo_check (x,y,dtr)
```

```python
algo_check (x,y,rfr)
```

```python
algo_check(x,y,svr)
```

```python
# it seems that rfr is the best algo among all the above checked algos. let's verify its overfitting.
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=99)
rfr.fit(x_train,y_train)
y_pred=rfr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
cross_val_score(rfr,x,y,cv=5,scoring='r2')
```

```python
# hence, the model is not overfitted.
```

```python
from sklearn.model_selection import GridSearchCV
from  sklearn.linear_model import Lasso,LassoCV
```

```python
def lasso_check(x_train,y_train,x_test,y_test):
    lasscv = LassoCV(alphas=[0.001,0.01,0.1,1,10],max_iter = 100, normalize =True)
    lasscv.fit(x_train,y_train)
    alp=lasscv.alpha_
    lasso_reg =Lasso(alp)
    lasso_reg.fit(x_train,y_train)
    score = lasso_reg.score(x_test,y_test)
    print(f"lasso accuracy is {score}")
```
